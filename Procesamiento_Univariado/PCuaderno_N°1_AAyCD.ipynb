# **Marco conceptual**

## **Información general acerca de la obtención de los datos**
El Sistema de Registro Civil y Estadísticas Vitales está orientado a recoger información sobre algunos hechos que ocurren a
 lo largo de la vida de las personas, tales como nacimiento, matrimonio, divorcio, anulación, adopción, legitimación,
 reconocimiento y defunción. Dentro de este sistema se identifican dos subsistemas, registro civil y estadísticas vitales. En particular, el sistema de estadísticas vitales es un proceso que acopia información de los hechos vitales, las características pertinentes a los propios sucesos, para luego compilar,sistematizar, analizar, evaluar, presentar y difundir esos datos. Actualmente, el Sistema de Estadísticas Vitales en Colombia sólo incluye registros de nacimientos y defunciones fetales y no fetales. En este caso, nos centraremos específicamente en los registros de nacimientos ocurridos en la ciudad de Bogotá, correspondientes al añ0 2023.

## **Origen de los datos**

El origen de los datos proviene del Departamento Administrativo Nacional de Estadística (DANE), el cual es responsable de la recolección, procesamiento y difusión de las estadísticas vitales en Colombia, que incluye como ya se ha mencionado registros de nacimientos. A través de la Coordinación de Estadísticas Vitales, el DANE mantiene actualizadas estas estadísticas, asegurando su calidad y accesibilidad para los usuarios. Los datos se obtienen mediante un proceso metodológico riguroso que abarca desde la recolección de información hasta su difusión, garantizando la precisión y utilidad de los mismos para diversos fines estadísticos y administrativos.

## **Descripción de los datos**
El conjunto de datos incluye información de un total de 60351 observaciones, con 35 características iniciales de las cuales 8 son numéricas y 27 categóricas. Entre las variables disponibles se encuentran el sexo, peso y talla al nacer, mes y lugar de nacimiento, persona que atendió el parto, y el sitio donde ocurrió. Además, se registran datos relevantes de los padres, tales como edad, estado civil, nivel educativo, régimen de seguridad social, y lugar y área de residencia tanto de la madre como del padre, entre otros.

Parte de lo descrito anteriormente fue tomado de  https://microdatos.dane.gov.co/index.php/catalog/775/study-description. ## **Matriz de correlación, de varianza muestral y poblacional**

```python
# Lista de columnas numéricas iniciales
from tabulate import tabulate
columnas_numericas = df[['PESO_GRAMOS', 'TALLA_CENTIMETROS', 'NUM_CONSULTAS_PRENAT',
                      'EDAD_MADRE', 'NUM_HIJOS_NACIDOS_VIVOS', 'NUM_EMBARAZOS', 'EDAD_PADRE','TIEMPO_GESTACION']]
matriz_correlacion = columnas_numericas.corr()

# representar la matriz de correlación en una tabla
tabla_correlacion = tabulate(
    matriz_correlacion,
    headers=matriz_correlacion.columns,  # Encabezados de columnas
    tablefmt='fancy_grid',
    showindex=True
)
print(tabla_correlacion)
```

```python
plt.figure(figsize=(10, 8))
# Dibujar el heatmap
sns.heatmap(matriz_correlacion, annot=True, cmap="cool", fmt='.2f')
plt.title("Matriz de Correlación")
# Mostrar el gráfico
plt.show()
```

```python
matriz_covarianza_muestral = columnas_numericas.cov()
tabla_covarianza_muestral = tabulate(
    matriz_covarianza_muestral,
    headers=matriz_covarianza_muestral.columns,  # Encabezados de columnas
    tablefmt='fancy_grid',
    showindex=True  # Mostrar índice (filas)
)
print(tabla_covarianza_muestral)
```

```python
# --- Matriz de varianzas y covarianzas poblacionales ---
n = len(columnas_numericas)
matriz_covarianza_poblacional = matriz_covarianza_muestral * (n - 1) / n
tabla_matriz_covarianza_poblacional = tabulate(
    matriz_covarianza_poblacional ,
    headers=matriz_covarianza_poblacional.columns,
    tablefmt='fancy_grid',
    showindex=True
)
print(tabla_matriz_covarianza_poblacional)
```

```python
# Función para clasificar valores según Tukey
def clasificar_valor(valor, q1, q3, iqr):
    if pd.isna(valor):
        return "F"  # Faltante
    elif valor < (q1 - 1.5 * iqr):
        return "AI"  # Atípico Inferior
    elif valor > (q3 + 1.5 * iqr):
        return "AS"  # Atípico Superior
    else:
        return "T"  # Típico

# Variables cuantitativas
variables_cuantitativas = [
    "PESO_GRAMOS", "TALLA_CENTIMETROS", "TIEMPO_GESTACION",
    "NUM_CONSULTAS_PRENAT", "EDAD_MADRE", "NUM_HIJOS_NACIDOS_VIVOS",
    "NUM_EMBARAZOS", "EDAD_PADRE"
]

# Crear nuevas variables categóricas
for variable in variables_cuantitativas:
    q1 = df[variable].quantile(0.25)
    q3 = df[variable].quantile(0.75)
    iqr = q3 - q1

    nueva_variable = f"{variable}_categoria"
    df[nueva_variable] = df[variable].apply(clasificar_valor, args=(q1, q3, iqr))

```

```python
caracterizacion_atipicos = {}
for variable in variables_cuantitativas:
    nueva_variable = f"{variable}_categoria"

    conteos = df[nueva_variable].value_counts()

    # Extraer el número de atípicos inferiores y superiores
    num_atipicos_inferiores = conteos.get("AI", 0)  # 0 si no hay AI
    num_atipicos_superiores = conteos.get("AS", 0)  # 0 si no hay AS

    # Total de atípicos
    total_atipicos = num_atipicos_inferiores + num_atipicos_superiores

    caracterizacion_atipicos[variable] = {
        "Atípicos Inferiores": num_atipicos_inferiores,
        "Atípicos Superiores": num_atipicos_superiores,
        "Total Atípicos": total_atipicos
    }
df_caracterizacion = pd.DataFrame(caracterizacion_atipicos).T
df_caracterizacion
```
